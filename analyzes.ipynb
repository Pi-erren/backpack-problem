# Compare repair methods

### Fitness evolution: generic example

```python
# Parameters
np.random.seed(0)
population_size = 100
mutation_rate = 0.01
crossover_rate = 0.7
num_generations = 100
budgets = [200, 300, 250, 350, 400]
utilities = np.random.randint(10, 100, size=50)
costs = np.random.randint(1, 50, size=(50, 5))
elitism_rate = 0.1
num_instance = 30

mean_fitness_low_ratio, std_fitness_low_ratio, mean_diversity_low_ratio, std_diversity_low_ratio = run_experiments(
    population_size, mutation_rate, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_ratio_utility_cost', num_instance
)
 
mean_fitness_low_algorithm_1, std_fitness_low_algorithm_1, mean_diversity_low_algorithm_1, std_diversity_low_algorithm_1 = run_experiments(
    population_size, mutation_rate, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_algorithm_1', num_instance
)

# Graphics
plt.figure(figsize=(12, 10))
plt.subplot(2, 1, 1)
plt.plot(mean_fitness_low_ratio, label='Mean Fitness: utility/cost repair method')
plt.fill_between(range(num_generations), mean_fitness_low_ratio - std_fitness_low_ratio, mean_fitness_low_ratio + std_fitness_low_ratio, alpha=0.2)
plt.plot(mean_fitness_low_algorithm_1, label='Mean Fitness: subject repair method')
plt.fill_between(range(num_generations), mean_fitness_low_algorithm_1 - std_fitness_low_algorithm_1, mean_fitness_low_algorithm_1 + std_fitness_low_algorithm_1, alpha=0.2)
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.legend()
plt.title('Fitness Evolution')
```

### Fitness evolution: Same utilities, different costs

```python
# Parameters
population_size = 100
crossover_rate = 0.7
num_generations = 100
budgets = [100, 200, 150, 120, 130]
elitism_rate = 0.1
num_instances = 30
mutation_rate = 0.01

# Generate utilities with the same value for all objects
utilities = np.ones(50) * 50

# Generate random costs for each object
costs = np.random.rand(50, 5) * 100


mean_fitness_low_ratio, std_fitness_low_ratio, mean_diversity_low_ratio, std_diversity_low_ratio = run_experiments(
    population_size, mutation_rate, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_ratio_utility_cost', num_instances
)
 
mean_fitness_low_algorithm_1, std_fitness_low_algorithm_1, mean_diversity_low_algorithm_1, std_diversity_low_algorithm_1 = run_experiments(
    population_size, mutation_rate, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_algorithm_1', num_instances
)

# Graphics
plt.figure(figsize=(12, 10))
plt.subplot(2, 1, 1)
plt.plot(mean_fitness_low_ratio, label='Mean Fitness Low Mutation Rate (Ratio Repair)')
plt.fill_between(range(num_generations), mean_fitness_low_ratio - std_fitness_low_ratio, mean_fitness_low_ratio + std_fitness_low_ratio, alpha=0.2)
plt.plot(mean_fitness_low_algorithm_1, label='Mean Fitness Low Mutation Rate (Algorithm 1 Repair)')
plt.fill_between(range(num_generations), mean_fitness_low_algorithm_1 - std_fitness_low_algorithm_1, mean_fitness_low_algorithm_1 + std_fitness_low_algorithm_1, alpha=0.2)
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.legend()
plt.title('Fitness Evolution')
```

## Diversity evolution according to mutation rate

```python
# Parameters
population_size = 100
crossover_rate = 0.7
num_generations = 100
budgets = [100, 200, 150, 120, 130]
utilities = np.random.uniform(50, 100, size=50)
costs = np.random.uniform(1, 50, size=(50, 5))
elitism_rate = 0.1
num_instances = 30

# Run experiments with low mutation rate
mutation_rate = 0.01
mean_fitness_low_ratio, std_fitness_low_ratio, mean_diversity_low_ratio, std_diversity_low_ratio = run_experiments(
    population_size, mutation_rate, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_ratio_utility_cost', num_instances
)

mean_fitness_low_algorithm_1, std_fitness_low_algorithm_1, mean_diversity_low_algorithm_1, std_diversity_low_algorithm_1 = run_experiments(
    population_size, mutation_rate, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_algorithm_1', num_instances
)

# Run experiments with high mutation rate
mutation_rate_high = 0.1
mean_fitness_high_ratio, std_fitness_high_ratio, mean_diversity_high_ratio, std_diversity_high_ratio = run_experiments(
    population_size, mutation_rate_high, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_ratio_utility_cost', num_instances
)

mean_fitness_high_algorithm_1, std_fitness_high_algorithm_1, mean_diversity_high_algorithm_1, std_diversity_high_algorithm_1 = run_experiments(
    population_size, mutation_rate_high, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, 'repair_algorithm_1', num_instances
)

# Graphics
plt.figure(figsize=(12, 10))

# Fitness Evolution
plt.subplot(2, 1, 1)
plt.plot(mean_fitness_low_ratio, label='Mean Fitness Low Mutation Rate (Ratio Repair)')
plt.fill_between(range(num_generations), mean_fitness_low_ratio - std_fitness_low_ratio, mean_fitness_low_ratio + std_fitness_low_ratio, alpha=0.2)
plt.plot(mean_fitness_high_ratio, label='Mean Fitness High Mutation Rate (Ratio Repair)')
plt.fill_between(range(num_generations), mean_fitness_high_ratio - std_fitness_high_ratio, mean_fitness_high_ratio + std_fitness_high_ratio, alpha=0.2)
plt.plot(mean_fitness_low_algorithm_1, label='Mean Fitness Low Mutation Rate (Algorithm 1 Repair)')
plt.fill_between(range(num_generations), mean_fitness_low_algorithm_1 - std_fitness_low_algorithm_1, mean_fitness_low_algorithm_1 + std_fitness_low_algorithm_1, alpha=0.2)
plt.plot(mean_fitness_high_algorithm_1, label='Mean Fitness High Mutation Rate (Algorithm 1 Repair)')
plt.fill_between(range(num_generations), mean_fitness_high_algorithm_1 - std_fitness_high_algorithm_1, mean_fitness_high_algorithm_1 + std_fitness_high_algorithm_1, alpha=0.2)
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.legend()
plt.title('Fitness Evolution')

# Diversity Evolution
plt.subplot(2, 1, 2)
plt.plot(mean_diversity_low_ratio, label='Mean Diversity Low Mutation Rate (Ratio Repair)')
plt.fill_between(range(num_generations), mean_diversity_low_ratio - std_diversity_low_ratio, mean_diversity_low_ratio + std_diversity_low_ratio, alpha=0.2)
plt.plot(mean_diversity_high_ratio, label='Mean Diversity High Mutation Rate (Ratio Repair)')
plt.fill_between(range(num_generations), mean_diversity_high_ratio - std_diversity_high_ratio, mean_diversity_high_ratio + std_diversity_high_ratio, alpha=0.2)
plt.plot(mean_diversity_low_algorithm_1, label='Mean Diversity Low Mutation Rate (Algorithm 1 Repair)')
plt.fill_between(range(num_generations), mean_diversity_low_algorithm_1 - std_diversity_low_algorithm_1, mean_diversity_low_algorithm_1 + std_diversity_low_algorithm_1, alpha=0.2)
plt.plot(mean_diversity_high_algorithm_1, label='Mean Diversity High Mutation Rate (Algorithm 1 Repair)')
plt.fill_between(range(num_generations), mean_diversity_high_algorithm_1 - std_diversity_high_algorithm_1, mean_diversity_high_algorithm_1 + std_diversity_high_algorithm_1, alpha=0.2)
plt.xlabel('Generations')
plt.ylabel('Diversity')
plt.legend()
plt.title('Diversity Evolution')

plt.tight_layout()
plt.show()
```

## Computation time according to population size

```python
# Parameters
mutation_rate = 0.1
crossover_rate = 0.7
num_generations = 100
budgets = [100, 200, 150, 120, 130]
utilities = np.random.uniform(50, 100, size=50)
costs = np.random.uniform(1, 50, size=(50, 5))
elitism_rate = 0.1
num_instances = 10
population_sizes = [20, 40, 60, 100, 200]

# Compare time for different population sizes
avg_times_ratio, std_times_ratio, avg_times_algorithm_1, std_times_algorithm_1 = compare_population_size_computation_times(
    population_sizes, mutation_rate, crossover_rate, num_generations, budgets, utilities, costs, elitism_rate, num_instances
)

# Graphics
plt.figure(figsize=(10, 6))

plt.errorbar(population_sizes, avg_times_ratio, yerr=std_times_ratio, fmt='-o', label='Repair Ratio Utility/Cost')
plt.errorbar(population_sizes, avg_times_algorithm_1, yerr=std_times_algorithm_1, fmt='-o', label='Repair Algorithm 1')

plt.title('Comparison of Computation Time for Different Population Sizes')
plt.xlabel('Population Size')
plt.ylabel('Time (seconds)')
plt.xticks(population_sizes)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
